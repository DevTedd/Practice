```python
data = df[['INWARDS_CO','UW_YEAR', 'LINE_OF_BUSINESS', 'CEDANT',
           'GBP','PREM_PF_IN','PREM_PF_OUT','CLAIMS_PAID',
           'CLAIM_PF_IN','CLAIM_PF_OUT',
           'CASH_LOSS_REFUND', 'PROF.COMM','TAX_ON_PREM',
           'MGNT_CHARGES', 'OTHER_CHARGE', 'OV_COMMISSION',
           'INT_ON_DEPOSITS']]
data.info()
```

```python
#First step is to check if we have any NaN values that can affect later steps
data.isna().sum().sum()
#Deleting rows isnt an option so the next step would be to replace the missing ones with '0'
# However we will do this for specific variables as not all collumns would benefit, i.e 
data.replace(np.nan,0)
data.fillna(0)
data.isna().sum().sum()
```

```python
data.isna().sum().sum()
```

```python
#We then edit the columns that are objects into numeric. First step is to remove the commas
data['PREM_PF_IN'] = data['PREM_PF_IN'].str.replace(',','')
data['PREM_PF_OUT'] = data['PREM_PF_OUT'].str.replace(',','')
data['CLAIMS_PAID'] = data['CLAIMS_PAID'].str.replace(',','')
data['CLAIM_PF_IN'] = data['CLAIM_PF_IN'].str.replace(',','')
data['CLAIM_PF_OUT'] = data['CLAIM_PF_OUT'].str.replace(',','')


data['PROF.COMM'] = data['PROF.COMM'].str.replace(',','')
data['TAX_ON_PREM'] = data['TAX_ON_PREM'].str.replace(',','')
data['MGNT_CHARGES'] = data['MGNT_CHARGES'].str.replace(',','')
data['OTHER_CHARGE']  = data['OTHER_CHARGE'].str.replace(',','')
data['OV_COMMISSION']  = data['OV_COMMISSION'].str.replace(',','')
data['INT_ON_DEPOSITS']  = data['INT_ON_DEPOSITS'].str.replace(',','')

data[['PROF.COMM', 'TAX_ON_PREM','MGNT_CHARGES', 'OTHER_CHARGE',
         'OV_COMMISSION','INT_ON_DEPOSITS']] = data[['PROF.COMM', 'TAX_ON_PREM','MGNT_CHARGES','OTHER_CHARGE',
                    'OV_COMMISSION','INT_ON_DEPOSITS']].apply(pd.to_numeric)
#The next step is to convert to numerict data type 

data[['PREM_PF_IN', 'PREM_PF_OUT']] = data[['PREM_PF_IN', 'PREM_PF_OUT']].apply(pd.to_numeric)
data[['CLAIMS_PAID', 'CLAIM_PF_IN','CLAIM_PF_OUT']] = data[['CLAIMS_PAID', 'CLAIM_PF_IN','CLAIM_PF_OUT']].apply(pd.to_numeric)
data.isna().sum().sum()
data.info()
```

```python
#Confirming if all looks good
data.tail(30)

data.fillna(0,inplace=True)
data.isna().sum().sum()

data.tail(5)
```

```python
data.isna().sum().sum()
print('No more null or missing values!')
```

We have our sections split and cleaning functioning well. Next phase is doing calculations

# Second Step

```python
data.head(5)
```

```python
#We'll be calculating the final premium, claims and expenses calculations into new colums

#Making the calculated premium column
data['F_PREM'] = data['GBP'] + data['PREM_PF_IN'] - data['PREM_PF_OUT']
data['F_PREM'] = data['F_PREM'].abs()

#Creating the calculated claims column 
data['F_CLAIM'] = data['CLAIMS_PAID'] - data['CLAIM_PF_IN'] +  data['CLAIM_PF_OUT'] - data['CASH_LOSS_REFUND']
data['F_CLAIM'] = data['F_CLAIM'].abs()

#Followed by expenses 

data['F_EXPENSE'] = data['PROF.COMM'] + data['TAX_ON_PREM']+data['MGNT_CHARGES']+data['OTHER_CHARGE']+data['OV_COMMISSION']+data['INT_ON_DEPOSITS']
data['F_EXPENSE'] = data['F_EXPENSE'].abs()
```

# Aggregate Functions

**The 3 cells below are transformation aggregating the values for claim. premiums and expenses together. Grouped by year and ID number**

```python
#Grouping Prems
worked_data = data.groupby(['INWARDS_CO','UW_YEAR','LINE_OF_BUSINESS'], as_index=False)['F_PREM'].agg("sum")
#data["new"] = new_column.reset_index(level=0, drop=True)
worked_data
```

```python
#Grouping Claims
new_column = data.groupby(['INWARDS_CO','UW_YEAR', 'LINE_OF_BUSINESS'], as_index=False)['F_CLAIM'].agg("sum")
#data["new"] = new_column.reset_index(level=0, drop=True)
worked_data['F_CLAIM'] = new_column['F_CLAIM']
```

```python
#Grouping expenses
new_column = data.groupby(['INWARDS_CO','UW_YEAR', 'LINE_OF_BUSINESS'], as_index=False)['F_EXPENSE'].agg("sum")
#data["new"] = new_column.reset_index(level=0, drop=True)
worked_data['F_EXPENSE'] = new_column['F_EXPENSE']
```

Now That its been grouped by year, we will go a step further a group it by contract(INWARDS_CO)

```python
worked_data['cost'] = worked_data['F_CLAIM'] + worked_data['F_EXPENSE']
```

```python
new_column = worked_data.groupby(['INWARDS_CO','LINE_OF_BUSINESS'], as_index=False)['F_PREM'].agg("sum")
new_column2 = worked_data.groupby(['INWARDS_CO'], as_index=False)['cost'].agg("sum")

new_column
```

```python
df2 = new_column
df2['Cost'] = new_column2['cost']
df2['year'] = worked_data['UW_YEAR']
df2
```

```python
df2['%'] = df2['Cost'] / df2['F_PREM']
df2
```

```python
df2.info()
#Lets change the LOB column into a string and
# use it to create the profit code
```

# We now have all the required varaible we need togther. We will now create the column that holds the profitabilty matrix

The range for the profitability matrix needs to be editable. We can use conditions to make this easier

```python
#This cell holds the condition for the profitability matrix

conditions = [
    (df2['%'] > 1),
    (df2['%'] <= 1) & (df2['%'] > 0.9),
    (df2['%'] < 0.9)
]

values = ['Onerous', 'Acceptable','Profitable']

v_codes = ['O','U','P']
```

```python
#df2['Profit_Code'] = np.select(conditions,values)
df2['Profit_Code'] = np.select(conditions,v_codes)
df2.head()
```

# Profitability Grouping

```python
df2[['LINE_OF_BUSINESS', 'year','%']] = df2[['LINE_OF_BUSINESS', 'year','%']].astype('string')
df2['Profit_Code'] = df2['LINE_OF_BUSINESS'] + '_' +df2['year'] + '_' + df2['Profit_Code']
df2.info()
```
